```python
# Class & Obj Attributes
class University:
    univ_name = "University of South Dakota" # Class Attribute
    univ_city = "Vermillion"              # Class Attribute

    def __init__(self, name, major):
        self.name = name  # Obj Attribute
        self.major = major # Obj Attribute


stu1 = University("Chenchu", "AI")
print(f"Student name : {stu1.name}, Major: {stu1.major}, University: {stu1.univ_name}")

```

```python
class Student:
    univ_namme = "University of South Dakota"  # Class Attribute

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def welcome_student(self):
        print(f"Warm Welcome to {self.univ_namme}, {self.name}!")

    def student_details(self):
        print(f"Student name : {self.name}, Age: {self.age}!.")

s1 = Student("Chenchu", 28)

s1.welcome_student()
s1.student_details()

```

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def average_marks(self):
        avg = sum(self.marks)/len(self.marks)
        print(f"The average score of {self.name} is {avg}")
    

    
s1 = Student("Chenchu", [99, 98, 99])
s1.average_marks()

s1.name = "Sai Krishna Yadav" # Name will be updated
s1.average_marks()

```

```python
class Apartment:
    apt_name = "MAA Eastside" # Class Attribute
    apt_city = "Richardson" # Class Attribute

    def __init__(self, name, apt_num, origin, move_in_date, move_out_date):
        self.name = name
        self.apt_num = apt_num
        self.origin = origin
        self.move_in_date = move_in_date
        self.move_out_date = move_out_date

    def moveIn(self):
        print(f"{self.name} has moved into apartment number {self.apt_num} on {self.move_in_date}.")

    def moveOut(self):
        print(f"{self.name} expected to moved out from apartment number {self.apt_num} on {self.move_out_date}.")


tenet1 = Apartment("Chenchu", 407, "India", "2025-07-01", "2027-06-30")
tenet2 = Apartment("Rajia", 408, "USA", "2024-09-15", "2026-09-14")

tenet1.moveIn()
tenet1.moveOut()

tenet2.moveIn()
tenet2.moveOut()

tenet2.name = "Sai Krishna Yadav"
tenet2.moveIn()
```

```python
# Create account class with 2 attributes - balance, acc_no. Create methods for debit and credit and printing the balace

class Account:

    def __init__(self, bal, acc_no):
        self.balance = bal
        self.acc_no = acc_no

    # Debit method
    def debit(self, amount):
        self.balance -= amount
        print(f"The updated balance is {self.balance} after withdrawing {amount}")
        self.get_balance()
    # Credit Method
    def credit(self, amount):
        self.balance += amount
        print(f"The updated balance is {self.balance} after crediting {amount}")
        self.get_balance()

    def get_balance(self):
        print(f"The updated balance is {self.balance}")


cust1 = Account(10000, 1234)

cust1.get_balance()
cust1.credit(1000)
cust1.credit(5000)
```
