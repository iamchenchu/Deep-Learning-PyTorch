### Pytorch Imp Concepts


```python
# Creating two random tensors and performing basic operations

import torch 

a = torch.rand(4, 3)
b = torch.rand(4, 3)

print("Tensor a:", a)
print("Tensor b:", b)
```

```python
print(f"a={a}, \n b={b}")
```

```python
add = a + b
sub = a - b
mul = a * b

print("Addition:\n", add)
print("Subtraction:\n", sub)
print("Multiplication:\n", mul)
```

```python
# Matrix multiplication, note the use of transpose

a_transpose = a.T
print(f"Shape -> a: {a.shape}")
print(f"Shape -> a_transpose: {a_transpose.shape}")
```

```python
dot_AB = a @ b.T
print(dot_AB)
```

```python
dot_BA = a.T @ b
print(dot_BA)
```

```python
# Broadcasting rules in PyTorch which allow operations on tensors of different shapes

c = torch.rand(4, 1)
print(f"Tensor c:\n{c}")
```
